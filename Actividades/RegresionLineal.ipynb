```python
#Actividad 3. Solución de problemas. Mauricio Sifuentes Rincón 2007693. Hora clase N4
#Modelo que hace la predicción de la calificación de un grupo de estudiantes
#tomando en cuenta el número de horas dedicadas a estudiar.
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

#El conjunto de datos se generará con la función “random” de Numpy de la siguiente manera:
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2

#se utilizará la función “train_test_split” de la librería scikit-learn
#para generar los conjuntos de datos de entrenamiento y datos de prueba.
train_data, test_data, train_label, test_label,= train_test_split(horas_estudio, calificaciones, test_size=0.20, random_state=0)

#Se entrenará el modelo de regresión lineal con los datos de entrenamiento con la función “fit” de la librería scikit-learn.
regressor = LinearRegression()
regressor.fit(train_data, train_label)

#Realizar predicciones con el conjunto de datos de prueba (train_data) y graficar el modelo de regresión lineal con estas predicciones.
prediction=regressor.predict(test_data)

#Grafica
plt.scatter(test_data, test_label, color='purple', label='Datos reales prueba')
plt.plot(test_data, prediction, color='orange', label='Regresion lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de regresion lineal')
plt.legend()
plt.show()

#Se utilizará la función “mean_squared_error” de scikit-learn para calcular el error cuadrático medio,
#que será la métrica a utilizar para evaluar el desempeño del modelo durante la prueba.
error=mean_squared_error(test_label, prediction)
print(f'El error cuadratico medio es de:{error}')

```
